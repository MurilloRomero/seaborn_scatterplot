```python
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.shape
```

```python
penguins.columns
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm');
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species');
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', palette='mako');
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', s=20);
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', size='sex');
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', size='sex', sizes=[40,100]);
```

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', style='sex');
```

```python
markers = {'Male':'X', 'Female':'s'}
sns.scatterplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='species', style='sex', markers=markers);
```

```python
sns.scatterplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='island', alpha=0.6);
```

```python
sns.scatterplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='sex', edgecolor='black');
```
